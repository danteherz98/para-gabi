```python
import os

carpeta = "/Users/d.s/Desktop/project_gabi/img/"  # cambia esto por la ruta real
prefijo = "gabi"

# Extensiones consideradas como im√°genes
extensiones = (".jpg", ".jpeg", ".png", ".gif", ".bmp", ".tiff", ".heic")

contador = 0

for archivo in os.listdir(carpeta):
    if archivo.lower().endswith(extensiones):
        extension = os.path.splitext(archivo)[1]
        nuevo_nombre = f"{prefijo}{contador}{extension}"
        
        ruta_vieja = os.path.join(carpeta, archivo)
        ruta_nueva = os.path.join(carpeta, nuevo_nombre)
        
        os.rename(ruta_vieja, ruta_nueva)
        contador += 1

print("Renombrado completado.")
```
